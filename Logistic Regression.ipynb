logistics regression

logistics regression is uesd for binary classfication where we used sigmoid function that takes input as independent variables, ans produces as a probability value between 0 &1

Types of Logistics Regression

1. binomial : there can be only two possible types of the dependent variables such 0 or 1, pass or fail, etc.
2. multinomial: there can be 3 or more possible unorderd types of the dependent variables such as "cat","dogs",etc
3. ordinal:  there can be 3 or more possible ordered types of dependent variables such as"low","medium","high",etc
